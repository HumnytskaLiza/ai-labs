Завдання 1: На вхід функція приймає рядок s. Функція повинна повернути True якщо рядок є паліндромом та False – в протилежному випадку.

Наприклад:

*   `is_palindrome(s='anade') -> False`
*   `is_palindrome(s='ana') -> True`




````python
def is_palindrome(s: str) -> bool:

  isPalindrome = True

  punctuation_list = [',', '.', ':', ";", "!", "?", " "]
  s = s.lower()
  for p in punctuation_list:
      if p in s:
          s = s.replace(p, "")

  j = len(s) - 1

  for i in s:
    if i == s[j]:
      isPalindrome = True
      j -= 1
    else:
      isPalindrome = False
      break
    
  return isPalindrome

print(is_palindrome(s='noon'))
print(is_palindrome(s='yelyzaveta'))
print(is_palindrome(s='palindrome'))
print(is_palindrome(s='Never, odd or even'))
````

Завдання 2: Розв’язати квадратне рівняння $ax^2+bx + c = 0$.  Напишіть функцію,яка приймає параметри квадратного рівняння, а повертає одне рішення або декілька.

Наприклад:  


```
solve_quad(a=1, b=-2, c=1) -> 1  
solve_quad(a=1, b=0, c=-1) -> (-1, 1)  
solve_quad(a=0, b=2, c=-1) -> 0.5
```







````python
import math

def solve_quad(a: float, b: float, c: float):

  discriminant = (b**2) - (4*a*c)

  if a == 0:
      x1 = -c/b
      return x1

  elif discriminant < 0:
    return "Discriminant < 0. No solutions."

  elif discriminant == 0:
      x1 = (-b) / (2*a)
      return x1

  else:
    x1 = (-b + math.sqrt(discriminant)) / (2*a)
    x2 = (-b - math.sqrt(discriminant)) / (2*a)
    return (x1, x2)
  
print(solve_quad(1, -2, 1))
print(solve_quad(2, 9, -5))
print(solve_quad(4, 20, 35))
print(solve_quad(4, 20, 25))
````

Завдання 3: Реалізація функції merge. На вхід функція приймає 2 відсортованих масиви - `a` і `b`. Функція повинна повернути один відсортований масив, який містить всі елементи масивів `a` і `b`.

Наприклад: `merge([1, 3, 3, 5], [3, 5, 5, 7]) -> [1, 3, 3, 3, 5, 5, 5, 7]`

````python
from typing import List

def merge(a: List[int], b: List[int]) -> List[int]:
  
  merged_list = []
  i = 0
  j = 0

  while i < len(a) and j < len(b):
    if a[i] < b[j]:
      merged_list.append(a[i])
      i += 1
    elif a[i] > b[j]:
      merged_list.append(b[j])
      j += 1
    else:
      merged_list.append(a[i])
      merged_list.append(b[j])
      j += 1
      i += 1
  
  for el_a in range (i, len(a)):
    merged_list.append(a[el_a])
  
  for el_b in range (j, len(b)):
    merged_list.append(b[el_b])

  return merged_list

print(merge([1, 3, 7, 9, 12, 13, 20], [3, 5, 5, 7, 8]))
print(merge([1, 3, 3, 5], [3, 5, 5, 7]))
````

Завдання 4: Реалізація алгоритму [MergeSort](https://ru.wikipedia.org/wiki/%D0%A1%D0%BE%D1%80%D1%82%D0%B8%D1%80%D0%BE%D0%B2%D0%BA%D0%B0_%D1%81%D0%BB%D0%B8%D1%8F%D0%BD%D0%B8%D0%B5%D0%BC)..

Наприклад: `MergeSort([3, 1, 3, 5]) -> [1, 3, 3, 5]`

````python
from typing import List

def MergeSort(L: List[int]) -> List[int]:
  
  if len(L) <= 1:
    return L
  
  else: 
    sorted_list = []
    i = 0
    j = 0

    l_side = MergeSort(L[:len(L)//2])
    r_side = MergeSort(L[len(L)//2:])

    while i < len(l_side) and j < len(r_side):
      if l_side[i] < r_side[j]:
        sorted_list.append(l_side[i])
        i += 1
      elif l_side[i] > r_side[j]:
        sorted_list.append(r_side[j])
        j += 1
      else:
        sorted_list.append(l_side[i])
        sorted_list.append(r_side[j])
        j += 1
        i += 1

    for l_ind in range (i, len(l_side)):
      sorted_list.append(l_side[l_ind])
  
    for r_ind in range (j, len(r_side)):
      sorted_list.append(r_side[r_ind])

  return sorted_list
  
print(MergeSort([8, 3, 9, 7, 2, 10, -11, 0, 13, 13, 31, 0, 2]))
print(MergeSort([0]))
````

Завдання 8: Написати функцію яка приймає на вхід двовимірний масив, а повертає масив з діагональними елементами цієї матриці.

Наприклад: `[[1, 2], [3, 4]] -> [1 4]`

````python
def get_diag(X):
    diagonal_elems = np.diagonal(X)
    return diagonal_elems

Arr1 = np.array([[1, 7], [3, 0]], dtype = np.float64)
print(get_diag(Arr1))
````

Завдання 9: Написати функцію яка приймає на вхід двомірний масив, а повертає масив елементів які знаходяться на другорядній діагоналі. Приклад: `[[1, 2], [3, 4]] -> [2 3]`

````python
def sec_diag(X):
    reversed_arr = np.fliplr(X)
    diagonal_elems = np.diagonal(reversed_arr)
    return diagonal_elems

Arr1 = np.array([[1, 0, 7], [3, 0, 8], [2, 2, 9]], dtype = np.float64)
print(sec_diag(Arr1))
````

Завдання 10: Кого більше - чоловіків чи жінок?

````python
df_all_people = dataset[["Sex"]]

female = df_all_people.value_counts()['female']
male = df_all_people.value_counts()['male']

if female > male:
    print(f"There are more women ({female})")
else:
    print(f"There are more men ({male})")
````
